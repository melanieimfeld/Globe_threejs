```python
import pandas as pd
```

```python
data_out = pd.read_csv("WtoData_cleaned1.csv")
data_in = pd.read_csv("WtoData_cleaned2.csv")
data_in
```

```python
centroids = pd.read_csv("country_centroids.csv")
centroids = centroids[["Longitude", "Latitude","name_long"]]
```

```python
outflows = data_out[(data_out["Product/Sector"] == "Financial services") & (data_out["Reporting Economy"] != "European Union")]
outflows = outflows.replace('Hong Kong, China','Hong Kong')
outflows = outflows.replace('Russian Federation','Russia')
len(outflows)
```

```python
#pd.merge(restaurant_ids_dataframe, restaurant_review_frame, on='business_id', how='outer')
result = pd.merge(left=outflows, right=centroids,left_on="Partner Economy", right_on="name_long", how='left')
```

```python
len(result)
```

```python
outflows = result[["Reporting Economy","Partner Economy","Longitude","Latitude","Value"]]
outflows
```

```python
inflows = data_in[(data_in["Product/Sector"] == "Financial services") & (data_in["Reporting Economy"] != "European Union")]
```

```python
resultI = pd.merge(left=inflows, right=centroids,left_on="Reporting Economy", right_on="name_long", how='left')
```

```python
inflows = resultI[["Reporting Economy","Partner Economy","Longitude","Latitude","Value"]]
```

```python
inflows
```

```python
merged = pd.concat([inflows, outflows])
```

```python
merged.to_json("tradeflows.json", orient="records")
```
